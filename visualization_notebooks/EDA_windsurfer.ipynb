```python
def select_player_names(club):
    players_in_club = players.player_id.isin(df_stats[df_stats.club == club].player_id)
    player_names_dropdown.options =  [name for name in players[players_in_club].Name_ID.sort_values()]                                      
    player_names_dropdown.value = (player_names_dropdown.options[0],) 
    player_names_dropdown.description = 'Select players name'

    
player_names_dropdown = widgets.SelectMultiple()
select_player_names('Real Madrid')
```

```python
def view_image(player_name):
    data=[]
    
    for name in player_name:
        df = prices[prices.Name_ID == name]
        trace = go.Scatter(x=df.Date,
                           y=df.price,
                           name = name
                          )
        data.append(trace)
    
    layout = dict(
        title='Player price: ' + ', '.join(player_name),
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1,
                         label='1m',
                         step='month',
                         stepmode='backward'),
                    dict(count=6,
                         label='6m',
                         step='month',
                         stepmode='backward'),
                    dict(count=1,
                        label='2017 Only',
                        step='year',
                        stepmode='todate'),
                    dict(step='all')
                ])
            ),
            rangeslider=dict(thickness = 0.1),
            type='date'
        )
    )

    fig = dict(data=data, layout=layout)
    iplot(fig)
```

```python
i = widgets.interactive(select_player_names, club = club_names_dropdown)
j = widgets.interactive(view_image, player_name = player_names_dropdown)
display(i)
display(j)
# You can select multiple items
```

```python
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs
```

```python
# код для отрисовки графиков
def tsplot(df, lags=None, figsize=(12, 7), style='bmh'):
   
    if not isinstance(df, pd.Series):
        y = df['price']
        y.index = df.Date
    else:
         y = df
            
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        

        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.5)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.5)

        print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(y)[1])

        plt.tight_layout()
    return 
     
    
```

## Анализ динамики количества игроков в "системе" и ее влияние на цену 

```python
# Задаем временной ряд, для оси х
start_date = '2016-12-07'
end_date = '2017-04-01'
rng = pd.date_range(start_date, end_date)
```

```python
stats_w_last_price.added_on = pd.to_datetime(stats_w_last_price.added_on)


# число добавленых карточек по датам
no_players_added = stats_w_last_price.groupby('added_on')['player_id'].count()

#общее число игроков в этот день
no_of_players = no_players_added.cumsum()[rng].fillna(method='ffill')

plt.figure(figsize(20,12))
fig, ax1 = plt.subplots()


# выберем 100 случайных игроков (которые в игре с самого начала) и с ценами от  1000  до 50000
added_on_start = (stats_w_last_price.added_on <= pd.to_datetime('2016-09-23'))
middle_prices = (stats_w_last_price.price.isin(range(1000,50000)))

#расскоментируйте эту строку если хотите получить 100 случайно отобраных игроков 
#rand_id = stats_w_last_price[middle_prices & added_on_start]['player_id'].sample(100, axis=0)

rand_id = stats_w_last_price[middle_prices & added_on_start]['player_id']

ax1.set_ylabel('price', color='b')
ax1.tick_params('y', colors='b')        

# график средней цены
mean_price = prices[prices.player_id.isin(rand_id)].groupby('Date')['price'].mean()[rng]
mean_price.plot(x = rng, label = 'mean price', ax = ax1, color = 'b', linewidth = 2.0, legend = True)    

# график отдельных цен - закомментируйте если мешает анализу
#for i in rand_id:
#   prices[(prices.player_id == i)  & (prices.Date.isin(rng))].plot(x = rng, y = 'price', \
#                          ax = ax1, color = 'k', linewidth = 0.45, legend = False, logy = False)



#график количества добавленых игроков
ax2 = ax1.twinx()
ax2.set_ylabel('Количество игроков', color='r')
ax2.tick_params('y', colors='r')

no_of_players.plot(color ='r', label = "количество игроков", ax = ax2, linewidth = 2.0)


corr, p = scs.pearsonr(no_of_players, mean_price)

print("Pearson correlation {}, p=value: {}".format(corr, p))
#fig.tight_layout()
plt.show()
```


Можно заметить что во время выпуска новых игроков цена на старых проседает. Но уровень значимости этого проседания требует дальнейшего изучения. В статистику включены игроки с последними ценами из диапазона 1000-50000.
